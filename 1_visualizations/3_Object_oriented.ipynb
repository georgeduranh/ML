```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x **2
```

```python
fig = plt.figure()
#Axes = graphs
axes = fig.add_axes([.1, .1, .5 , .9])
axes.plot(x,y, 'b')
fig.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([.1, .1, .8 , .9])
axes2 = fig.add_axes([.2, .55, .4 , .3])
axes.plot(x,y, 'b')
axes2.plot(y,x, 'r')
axes2.set_facecolor('gray')
fig.show()
```
