```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A', 'B', 'C'], y=[1,3,2])
plt.show()
```

```python
sns.set(style='darkgrid', palette="dark", font="Verdana", font_scale=1)
sns.barplot(x=['A', 'B', 'C'], y=[1,3,2])
plt.show()
```

##Common parameters

```python
tips = sns.load_dataset('tips')
tips
```

```python
#PArameters  https://cheatography.com/aggialavura/cheat-sheets/python-seaborn-imgs/
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde', legend=False)
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=True, palette="dark")
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', legend=True, palette="dark", alpha=0.5)
plt.show()
```

##Distribution

```python
sns.histplot(data=tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='dodge')
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=False, shade=True, bw_adjust=1)
plt.show()
plt.savefig("name.png")
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show()
```

```python
#general
sns.displot(data=tips, x='tip', hue='sex', kind='kde')
plt.show()
```

##Categorical variables

```python
sns.countplot(data=tips, x='day', hue='sex');
```

```python
sns.countplot(data=tips, y='day', hue='sex');
```

```python
plt.figure(figsize=(6,6))
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True);
```

```python
plt.figure(figsize=(6,6))
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True);
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True);
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True);
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True, color="0");
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True);
```

```python
sns.catplot(data=tips, x="day", hue="size", col="time", kind= "count" );
```

##Relation

```python
sns.pairplot(data=tips,hue='day')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size');
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
```

```python
markers = {"Lunch": "D", "Dinner" : "s"} 
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers);
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
```

```python

sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers);
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers, kind='scatter', col='time');
```

## Joinplot and pairplot

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True, marginal_kws=dict(bins=25, fill=True, multiple='dodge'));
```

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True);
```
