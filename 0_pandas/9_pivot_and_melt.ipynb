```python
import pandas as pd
import numpy as np
```

```python
df_books = pd.read_csv('https://static.platzi.com/media/public/uploads/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv', sep=",", header=0)
df_books.head(2)
```

```python
df_books.pivot_table(index='Author',columns='Genre',values='User Rating')
```

```python
df_books.pivot_table(index='Genre',columns='Year', values='User Rating',aggfunc='sum')
```

```python
#Melt
df_books[['Name','Genre']].head(5)
```

```python
df_books[['Name','Genre']].head(5).melt()
```

```python
df_books.melt(id_vars='Year',value_vars='Genre')
```
