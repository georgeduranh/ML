```python
import pandas as pd
df_books = pd.read_csv('https://static.platzi.com/media/public/uploads/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv')
df_books.head(2)
```

```python
grether_than_2016 = df_books['Year'] > 2016
```

```python
df_books[grether_than_2016]
```

```python
df_books[df_books['Year'] > 2016]
```

```python
df_books[ (df_books['Genre'] == 'Fiction') &  (df_books['Year'] > 2016) ]
```

```python
df_books[ (df_books['Genre'] == 'Fiction') &  ~(df_books['Year'] > 2016) ]
```
