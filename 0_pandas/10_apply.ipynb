```python
import pandas as pd
```

```python
df_books = pd.read_csv('https://static.platzi.com/media/public/uploads/bestsellers-with-categories_e591527f-ae45-4fa5-b0d1-d50142128fa6.csv')
df_books.head(2)
```

```python
def two_times(value):
    return value * 2
```

```python
df_books['User Rating'].apply(two_times)
```

```python
df_books['User Rating2'] =df_books['User Rating'].apply(two_times)
```

```python
df_books['User Rating2'] =df_books['User Rating'].apply(lambda x: x* 3)
```

```python
df_books['User Rating2']  = df_books.apply(lambda x: x['User Rating'] * 2 if x['Genre'] == 'Fiction' else x['User Rating'], axis = 1)
```
